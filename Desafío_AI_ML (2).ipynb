***RANDOM FOREST MODEL***

```python
# Importing RandomForestClassifier from sklearn.ensemble
# We will be further discussing about why Random Forest is in ensemble module of sklearn library
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200,criterion='gini',random_state=1)
```

```python
rfc.fit(x_Train, y_Train)
```

```python
# Este atributo tiene una medida de la importancia de cada feature
feature_importance = rfc.feature_importances_ 

print('Features: {}'.format(x_Train))
print('Importancia = {}'.format(feature_importance))
```

```python
rfc.predict_proba(x_Test)
```

***Predicciones***

```python
predictions = rfc.predict(x_Test)
```

```python
predictions.shape
```

```python
#Score del modelo en el conjunto de entrenamiento 
score_in_train = rfc.score(x_Train, y_Train)
```

```python
print(score_in_train)
```

```python
#Score del modelo en el conjunto de testeo
score_in_test = rfc.score(x_Test, y_Test)
```

```python
print(score_in_test)
```

```python
#evaluate model
# accuracy
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, f1_score, precision_score, recall_score
```

```python
#test model
accuracy_clf = accuracy_score(y_Test, predictions)
auc_roc1 = roc_auc_score(y_Test, predictions)
print("Accuracy score of the model is: {}".format(accuracy_clf))
print("roc_auc_score of the model1 is: {}".format(auc_roc1))
```

```python
#Precisión
precision = precision_score(y_Test, predictions)
print("Precisión: ", precision)

#Recall
recall = recall_score(y_Test, predictions)
print("Recall: ", recall)

#F1 Score
f1 = f1_score(y_Test, predictions)
print("F1 Score: ", f1)
```

```python
from sklearn.metrics import mean_absolute_error

predictionsa = rfc.predict(x_Test)
print("Mean Absolute Error: " + str(mean_absolute_error(predictionsa, y_Test)))
```

```python
confusion_matrix(y_Test, predictions)
```

```python
matrix = confusion_matrix(y_Test, predictions)
matrix = matrix.astype('float') / matrix.sum(axis=1)[:, np.newaxis]
```
